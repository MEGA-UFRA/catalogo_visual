```python
import pandas as pd
#import numpy as np
```

```python
tabela1 = tabelas[3]
display(tabela1)
```

```python
#adicionando um prefixo aos nomes de coluna
tabela1.columns = 'DEP_' + tabela1.columns
display(tabela1)
```

```python
#filtrando o resultado de uma tabela em forma de dataframe: [[]]
filtro1 = tabela1.iloc[[0]]
display(filtro1)
```

```python
filtro2 = tabela1.iloc[[1]]

filtro2.columns = ['PD-EDg','PA-EDg','PS-EDg','PM-EMg','TMDg','IPPg','STAYg','PE-365g','PE-450g','ESOBg','PSOBg','MSOBg','AOLg','ACABg','iABCZg']
filtro2.columns = 'AC_' + filtro2.columns
display(filtro2)
```

```python
filtro3 = tabela1.iloc[[2]]

filtro3.columns = ['PD-EDg','PA-EDg','PS-EDg','PM-EMg','TMDg','IPPg','STAYg','PE-365g','PE-450g','ESOBg','PSOBg','MSOBg','AOLg','ACABg','iABCZg']
filtro3.columns = 'DECA_' + filtro3.columns
display(filtro3)
```

```python
tabelas[0]
```

```python
#Tetando pegar o id do animal partindo da tabela[0]

tabelas[0].columns.values.tolist()#pegando apenas o nome da coluna da tabela[0] em forma de lista
#usando o slipt() para quebrar a string em 2 partes
vetor = tabelas[0].columns.values.tolist()[0].split(':::  ')#aqui tem ::: e dois espaços em branco
print(vetor)
vetor[1]
```

```python
id = tabelas[0].columns.values.tolist()[0].split(':::  ')[1]
filtro1.insert(0,"id", id, True)
filtro2.insert(0,"id", id, True)
filtro3.insert(0,"id", id, True)
display(filtro1)
display(filtro2)
display(filtro3)
```

```python
#add a coluna id do animal aos dataframes filtro1, filtro2 e  filtro3
#usar o id do animal como chave
#usar a funcao Merge() e juntar os dataframes filtro1, filtro2 e  filtro3, usando como chave o id do animal

#Fazendo o merge entre as tabelas DEP (filtro1) e AC (filtro2)
result_df = pd.merge(filtro1, filtro2, on='id')
display(result_df)
```

```python
#Fazendo o merge entre as tabelas DEP (filtro1) e AC (filtro2) já em result_df com DECA (filtro3)
result_df = pd.merge(result_df, filtro3, on='id')
display(result_df)
```
