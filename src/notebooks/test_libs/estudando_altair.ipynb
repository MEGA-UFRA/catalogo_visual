```python
import altair as alt
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Anderson Soares\Documents\GitHub\catalogo_visual\databases\PLANILHA_NELORE_ABCZ_GRUPO_2_80.csv',sep=';')
display(df) 
```

```python
df.dtypes
```

```python
alt.Chart(df).mark_point().encode(
    y='IABCZ',
    x='NOME',
    color = 'NOME'
).properties(
    width = 1000,
    height = 1200
)
```

```python
# PA-ED (peso ao ano- efeito direto); 
# GPD-ED (ganho em peso pÃ³s-desmama),  
# PE-450 (perÃ­metro escrotal ao sobreano),  
# ACAB (acabamento de carcaÃ§a),  
# Precocidade,  
# Stayability

alt.Chart(df).transform_window(
    index='count()'
).transform_fold(
    ['DEP_PA_ED', 'DEP_PE_365', 'DEP_STAY', 'DEP_PA_ED', 'DEP_TMD',  'DEP_ACAB', 'DEP_P','DEP_PE_450']
).mark_line().encode(
    x='key:N',
    y='value:N',
    color="IABCZ",
    #detail='index:N',
    opacity=alt.value(0.5),
    tooltip="NOME:N"
).properties(
    width = 800,
    height = 600
).interactive()
```

```python
grafico = alt.Chart(df).mark_circle().encode(
    alt.X(alt.repeat("column"), type='quantitative'),
    alt.Y(alt.repeat("row"), type='quantitative'),
    color='DEP_PE_365:Q',
    tooltip="NOME:N"
).properties(
    width=150,
    height=150
).repeat(
    row=['DEP_PE_365', 'DEP_STAY', 'DEP_PA_ED'],
    column=['DEP_ACAB', 'DEP_P','DEP_PE_450']
).interactive()

#configuracoes do slider
animais_slider = alt.binding_range(min=1, max=df['DEP_PE_365'].count(), step=1)
slider_selection = alt.selection_single(bind=animais_slider, fields=['DEP_PE_365'], name="Quant animais")

filter_animais = grafico.add_selection(
    slider_selection
).transform_filter(
    slider_selection
).properties(title="VisÃ£o geral dos dados | Slider de filtro")

filter_animais
```
